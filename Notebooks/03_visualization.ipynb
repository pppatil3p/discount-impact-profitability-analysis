```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../dataset/final_analysis_data.csv")
```

```python
df.columns
```

```python
profit_by_discount = (
    df.groupby('discount_bucket')['profit']
    .sum()
    .sort_index()
)

plt.figure(figsize=(8,5))
profit_by_discount.plot(kind='bar')
plt.title("Total Profit by Discount Bucket")
plt.xlabel("Discount Bucket")
plt.ylabel("Total Profit")
plt.show()
```

```python
loss_rate = (
    df.groupby('discount_bucket')['is_loss']
    .mean()
    .sort_index()
)

plt.figure(figsize=(8,5))
loss_rate.plot(kind='bar')
plt.title("Loss Rate by Discount Bucket")
plt.xlabel("Discount Bucket")
plt.ylabel("Loss Rate")
plt.show()
```

Insight:

-Loss rate increases significantly beyond higher discount ranges

-Indicates risk of aggressive discounting
